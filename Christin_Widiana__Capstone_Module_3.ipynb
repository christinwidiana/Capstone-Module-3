```python
# Fungsi untuk menghitung persentase outliers
def calculate_outlier_percentage(column):
    if column.dtype in [np.float64, np.int64]:
        z_scores = np.abs(stats.zscore(column.dropna()))  # Menghitung skor Z, mengabaikan nilai NaN
        return np.sum(z_scores > 3) / len(column) * 100  # Menghitung persentase outliers
    return 0

# Membuat DataFrame dengan informasi yang diinginkan
df_info = pd.DataFrame({
    'Features'           : df_model.columns.values,
    'Data Type'         : df_model.dtypes.values,
    'Null Value (%)'     : df_model.isna().mean().values * 100,
    'Unique Value'          : df_model.nunique().values,
    'Duplicate Data'         : df_model.duplicated().sum(),
    'Negative Value': [len(df_model[col][df_model[col] < 0]) / len(df_model) * 100
                           if col in df_model.select_dtypes(include=np.number).columns else 0
                           for col in df_model.columns],
    'Zero Value Numeric'    : [len(df_model[col][df_model[col] == 0]) / len(df_model) * 100
                           if col in df_model.select_dtypes(include=np.number).columns else 0
                           for col in df_model.columns],
    'Zero Value Object'    : [len(df_model[col][df_model[col] == '']) / len(df_model) * 100
                           if col in df_model.select_dtypes(include=np.object_).columns else 0
                           for col in df_model.columns],
    'Outlier(%)'        : [calculate_outlier_percentage(df_model[col]) for col in df_model.columns],
    'Unique Sample'     : [df_model[col].unique() for col in df_model.columns]
}).round(2)

df_info
```

## Matrik evaluasi

### Inisialisasi
Persiapkan fungsi yang dirasa perlu untuk matriks evaluasi
dan inisialisasi setiap atribut evaluasi

```python
# Fungsi untuk menghitung adjusted r2
def calculate_adjusted_r2(r2, n, p):
    adj_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
    return adj_r2

# Fungsi untuk menghitung MPSE
def mean_percentage_squared_error(y_true, y_pred):
    return np.mean(np.square((y_true - y_pred) / y_true)) * 100

# Fungsi untuk menghitung metrik evaluasi
def evaluate_model(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    msle = mean_squared_log_error(y_true, y_pred)
    mpse = mean_percentage_squared_error(y_true, y_pred)
    return mse, rmse, mae, r2, mape, msle, mpse

# Inisialisasi list untuk setiap metrik evaluasi
score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []
adjusted_r_squared = []

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

score_msle = []
nilai_mean_msle = []
nilai_std_msle = []

score_mpse = []
nilai_mean_mpse = []
nilai_std_mpse = []

training_times = []
```

## Evaluasi Model

```python
# Cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=28)

# Evaluasi setiap model
for model_name, model in models.items():
    start_time = time.time()

    estimator = Pipeline(steps=[
        ('preprocessing', preprocessor),
        ('scaler', scaler),
        ('model', model)
    ])

    try:
        # R-Square
        model_cv_r2 = cross_val_score(
            estimator,
            x_train,
            y_train,
            cv=kf,
            scoring='r2',
            error_score='raise'
        )

        score_r2.append(model_cv_r2)
        nilai_mean_r2.append(model_cv_r2.mean())
        nilai_std_r2.append(model_cv_r2.std())

        # Menghitung Adjusted R-squared
        n = len(y_train)  # Number of samples in the training set
        p = x_train.shape[1]  # Number of features in the training set
        adj_r2 = calculate_adjusted_r2(model_cv_r2.mean(), n, p)
        adjusted_r_squared.append(adj_r2)

        # Root Mean Squared Error
        model_cv_rmse = cross_val_score(
            estimator,
            x_train,
            y_train,
            cv=kf,
            scoring='neg_root_mean_squared_error',
            error_score='raise'
        )

        score_rmse.append(model_cv_rmse)
        nilai_mean_rmse.append(-model_cv_rmse.mean())
        nilai_std_rmse.append(model_cv_rmse.std())

       # Mean Absolute Error
        model_cv_mae = cross_val_score(
            estimator,
            x_train,
            y_train,
            cv=kf,
            scoring='neg_mean_absolute_error',
            error_score='raise'
        )

        score_mae.append(model_cv_mae)
        nilai_mean_mae.append(-model_cv_mae.mean())
        nilai_std_mae.append(model_cv_mae.std())

        # Mean Absolute Percentage Error
        model_cv_mape = cross_val_score(
            estimator,
            x_train,
            y_train,
            cv=kf,
            scoring='neg_mean_absolute_percentage_error',
            error_score='raise'
        )

        score_mape.append(model_cv_mape)
        nilai_mean_mape.append(-model_cv_mape.mean())
        nilai_std_mape.append(model_cv_mape.std())

        model_cv_msle = cross_val_score(
            estimator,
            x_train,
            y_train,
            cv=kf,
            scoring='neg_mean_squared_log_error',
            error_score='raise')

        score_msle.append(model_cv_msle)
        nilai_mean_msle.append(model_cv_msle.mean())
        nilai_std_msle.append(model_cv_msle.std())

        model_cv_mpse = cross_val_score(
            estimator,
            x_train,
            y_train,
            cv=kf,
            scoring='neg_mean_squared_error',
            error_score='raise')

        score_mpse.append(model_cv_mpse)
        nilai_mean_mpse.append(model_cv_mpse.mean())
        nilai_std_mpse.append(model_cv_mpse.std())

        # Hitung waktu evaluasi
        training_times.append(time.time() - start_time)

        print(f"Model {model_name} evaluated successfully.")
    except Exception as e:
        print(f"Error evaluating model {model_name}: {e}")
```

```python
# Data evaluasi model
data_evaluasi = {
    'Model': ['Linear Regression', 'KNN Regressor', 'DecisionTree Regressor', 'Random Forest Regressor', 'XGBoost Regressor', 'Gradient Boosting Regressor', 'Ridge Regression', 'Lasso Regression', 'Support Vector Regression'],
    'Mean_R2': nilai_mean_r2,
    'Std_R2': nilai_std_r2,
    'Adjusted_R2': adjusted_r_squared,
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE': nilai_std_rmse,
    'Mean_MAE': nilai_mean_mae,
    'Std_MAE': nilai_std_mae,
    'Mean_MAPE': nilai_mean_mape,
    'Std_MAPE': nilai_std_mape,
    'Validation_Time': training_times
}

# Mengonversi data evaluasi ke DataFrame
results_df = pd.DataFrame(data_evaluasi)

# Membulatkan nilai numerik hingga 3 desimal
results_df.iloc[:, 1:] = results_df.iloc[:, 1:].round(3)

# Mengurutkan data berdasarkan RMSE secara ascending
results_df = results_df.sort_values(by='Mean_RMSE', ascending=True)

# Menampilkan hasil menggunakan tabulate dengan Model rata kiri
print(tabulate(results_df, headers='keys', tablefmt='pretty', showindex=False, stralign='left', numalign='right'))
```

# **Feature Importance**

Untuk dapat mengetahui sebenarnya fitur apa saja yang sangat memengaruhi target (price), kita dapat mengeceknya melalui function feature_importances_.

```python
xgb = XGBRegressor(random_state=28)

# Membuat pipeline dengan preprocessor dan model
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', xgb)])

# Latih model menggunakan data training
pipeline.fit(x_train, y_train)

# Predict menggunakan data test
y_pred = pipeline.predict(x_test)

# Mendapatkan feature importances dari model XGBoost
# Perlu diketahui bahwa kita harus mengambil feature importances dari estimator XGBoost setelah preprocessing
xgb_model = pipeline.named_steps['model']
feature_importances = xgb_model.feature_importances_

# Mendapatkan nama-nama fitur setelah preprocessing
encoded_feature_names = pipeline.named_steps['preprocessor'].get_feature_names_out()
feature_importances_df = pd.DataFrame({'Feature': encoded_feature_names, 'Importance': feature_importances})

# Menampilkan feature importances dalam bentuk plot bar
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(12, 8))
plt.barh(feature_importances_df['Feature'], feature_importances_df['Importance'], color='blue')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance from XGBoost Model')
plt.gca().invert_yaxis()
plt.show()
```
